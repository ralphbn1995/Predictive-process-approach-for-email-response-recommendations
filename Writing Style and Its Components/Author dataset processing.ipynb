Dataset preprocessing

```python
with open('../Data/dic_allThreads.txt') as f:
    json_data = json.load(f)

f = open('../Data/writing_style_dataset.csv', 'w')
writer = csv.writer(f, delimiter='*')
row = ["text","author"]
writer.writerow(row)
import re

data= []
for x in range(1, 1000):
  first_key = list(json_data.values())[x]

  try:
    first_ = list(first_key.values())[1]
  except IndexError:
    first_ = list(first_key.values())[0]
  str_ = []

  x = first_[4].split("From:")
  r = 0

  for i in x:

    row = ["text","author"]
    for line in i.splitlines():
        if len(line.strip())> 150 and i.partition('\n')[0]:  
            author = i.partition('\n')[0].strip()
            
            
            
            a_string = author.split()[0]
            alphanumeric = [character for character in a_string if character.isalnum()]
            alphanumeric = "".join(alphanumeric)
            for k in line.split("\n"):
                final = " ".join(re.findall(r"[a-zA-Z0-9]+", k))
                print(final)
                
            row = [final, alphanumeric]
            
            if "X-cc:" not in row[1]:
                if "To:" not in row[1]:
                    writer.writerow(row)
                    r=r+1
            
            print('Author: ' +  alphanumeric)
            
```
